```python
from __future__ import division

import kwant
import scipy.sparse.linalg as sla

from math import pi, sqrt, tanh
from matplotlib import pyplot
```

```python
sin_30, cos_30 = (1 / 2, sqrt(3 / 2))
```

```python
graphene_a = kwant.lattice.general([(sqrt(3)/2, 3/2), (sqrt(3), 0)], [(0, 0), (sqrt(3)/2, 1/2)], norbs = 2)
graphene_b = kwant.lattice.general([(0, 1), (sqrt(3)/2, 1/2)], [(sqrt(3)/3, 0), (sqrt(3)/6, 1/2)], norbs = 2)
```

```python
def circle(pos):
    x, y = pos
    return x**2 + y**2 < 12
```

```python
syst = kwant.Builder()
syst[graphene_a.shape(circle, (0, 0))] = 0
syst[graphene_a.neighbors()]=1
syst[graphene_b.shape(circle, (0, 0))] = 0
syst[graphene_a.neighbors()]=1

#def family_colors(site):
#    return 0 if site.family == a else 1

# Plot the closed system without leads.
#kwant.plot(syst, site_color=family_colors, site_lw=0.1, colorbar=False);

kwant.plot(syst, site_symbol='o', site_size = 0.1, site_color = '0')
```
